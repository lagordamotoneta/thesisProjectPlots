```python
import pandas as pd
from datetime import datetime
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

import numpy as np
```

```python
stars=pd.read_csv('Stars/starsRange.csv')
cc=pd.read_csv('CC/CCRange.csv')

cc['Date'] = pd.to_datetime(cc['Date'], format='%d/%m/%Y %H:%M:%S')
cc.set_index('Date', inplace=True)
ccs=cc[['Size']]

stars['Date'] = pd.to_datetime(stars['Date'], format='%d/%m/%Y %H:%M:%S')
stars.set_index('Date', inplace=True)
starsDaily=stars[['InDegree']]

fig, ax = plt.subplots(figsize=(30,10))
ax.plot(ccs, label='Size of largest connected component')
ax.plot(starsDaily, label='In-Degree of star')
plt.rcParams['xtick.labelsize']=15
plt.rcParams['ytick.labelsize']=15
plt.xticks(rotation=45)

legend = ax.legend(loc='upper left', shadow=True, fontsize='x-large',prop={'size': 25})

# plt.suptitle('InDegree of identified star Post, accumulated', fontsize=40)

plt.show()
```
