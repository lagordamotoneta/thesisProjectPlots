```python
import pandas as pd
from datetime import datetime
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

import numpy as np
```

# Posts and Replies Graph derived from sample file for hourly Window Analysis

### Vertices and edges

```python


densityVandE=pd.read_csv('Density/densityWindowHrCOMPLETE.csv')
densityVandE['Date'] = pd.to_datetime(densityVandE['Date'], format='%d/%m/%Y %H:%M:%S')
densityVandE.set_index('Date', inplace=True)
densityVandE=densityVandE[['TotalVertices','TotalEdges']]
densityVandE.plot(figsize=(30,10),style=['-', '--','.-'])
plt.grid(linestyle='-')
plt.rcParams['xtick.labelsize']=25
plt.rcParams['ytick.labelsize']=25
plt.legend(loc='upper left',prop={'size': 25})
plt.suptitle('Posts and Replies Graph hourly window', fontsize=30)


plt.show()
```

```python
sns.lmplot(x='TotalVertices', y='TotalEdges', fit_reg=False, data=densityVandE);
plt.figsize=(30,40)
plt.rcParams['xtick.labelsize']=15
plt.rcParams['ytick.labelsize']=15
plt.suptitle('Correlation between Vertices and Edges', fontsize=20)
```

```python


densityVandE=pd.read_csv('Density/densityWindowHrCOMPLETE.csv')
densityVandE['Date'] = pd.to_datetime(densityVandE['Date'], format='%d/%m/%Y %H:%M:%S')
densityVandE.set_index('Date', inplace=True)
densityVandE=densityVandE[['TotalVertices','TotalEdges']]
densityVandE=densityVandE['2017-08-19': '2017-08-21']
densityVandE.plot(figsize=(30,10),style=['-', '--','.-'])
plt.grid(linestyle='-')
plt.rcParams['xtick.labelsize']=25
plt.rcParams['ytick.labelsize']=25
plt.legend(loc='upper left',prop={'size': 25})
plt.suptitle('Posts and Replies Graph 19 to 21 of August 2017 hourly window', fontsize=30)


plt.show()
```

# Density

```python


density=pd.read_csv('Density/densityWindowHrCOMPLETE.csv')
density['Date'] = pd.to_datetime(density['Date'], format='%d/%m/%Y %H:%M:%S')
density.set_index('Date', inplace=True)
density=density[['Density']]
density.plot(figsize=(30,10),style=['-', '--','.-'])
plt.grid(linestyle='-')
plt.rcParams['xtick.labelsize']=25
plt.rcParams['ytick.labelsize']=25
plt.legend(loc='upper left',prop={'size': 25})
plt.suptitle('Posts and Replies Graph hourly window', fontsize=30)

plt.yscale('log')

plt.show()
```
