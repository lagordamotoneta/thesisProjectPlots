```python
import pandas as pd
from datetime import datetime
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

import numpy as np
```

# Vertices and Edges Comments wise

```python
densityVandE=pd.read_csv('Density/densityWindow.csv')
densityVandE['Date'] = pd.to_datetime(densityVandE['Date'], format='%d/%m/%Y %H:%M:%S')
densityVandE.set_index('Date', inplace=True)
densityVandE=densityVandE[['TotalVertices','TotalEdges','Density']]
densityVandE.plot(figsize=(30,10),style=['-', '--','.-'])
```

# Vertices Comments wise

```python

verticesMonth = densityVandE['2016-09-01': '2016-09-30']
verticesMonth.TotalVertices.plot(title='# of Vertices September')
```

```python
verticesDays = densityVandE['2016-09-25': '2016-09-29']
verticesDays.TotalVertices.plot(title='# of transactions around peak')
```

```python
verticesHrs = densityVandE['2016-09-26 18:00': '2016-09-27 03:00']
verticesHrs.TotalVertices.plot(title='# of transactions around peak')
```
