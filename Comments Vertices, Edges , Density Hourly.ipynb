```python
import pandas as pd
from datetime import datetime
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

import numpy as np
```

# Posts and Replies Graph derived from sample file for hourly Window Analysis

### Vertices and edges

```python


densityVandE=pd.read_csv('Density/densityWindowHr.csv')
densityVandE['Date'] = pd.to_datetime(densityVandE['Date'], format='%d/%m/%Y %H:%M:%S')
densityVandE.set_index('Date', inplace=True)
densityVandE=densityVandE[['TotalVertices','TotalEdges']]
densityVandE.plot(figsize=(30,10),style=['-', '--','.-'])
plt.grid(linestyle='-')
plt.rcParams['xtick.labelsize']=25
plt.rcParams['ytick.labelsize']=25
plt.legend(loc='upper left',prop={'size': 25})
plt.suptitle('Posts and Replies Graph hourly window', fontsize=30)


plt.show()
```

```python
densityVandE.diff().plot(linewidth=2, fontsize=15,title='Remove trend')
```

# Zoom in for a month

```python

verticesMonth = densityVandE['2016-10-01': '2016-10-10']
plt.rcParams['xtick.labelsize']=15
plt.rcParams['ytick.labelsize']=15
verticesMonth.plot(title='Zoom in October')
```

# Zoom in 09-10 October

```python
verticesDays = densityVandE['2016-10-09': '2016-10-10']
plt.rcParams['xtick.labelsize']=15
plt.rcParams['ytick.labelsize']=15
verticesDays.plot(title='Zoom in 09-10 October')
```

# Zoom in 9 Oct 20:00 to 10 Oct 03:00

```python
verticesHrs = densityVandE['2016-10-09 20:00': '2016-10-10 03:00']
verticesHrs.plot(title='Zoom in 9 Oct 20:00 to 10 Oct 03:00')
```

# Density

```python


density=pd.read_csv('Density/densityTestRangeInNeig.csv')
density['Date'] = pd.to_datetime(density['Date'], format='%d/%m/%Y %H:%M:%S')
density.set_index('Date', inplace=True)
density=densityVandE[['Density']]
density.plot(figsize=(30,10),style=['-', '--','.-'])
plt.grid(linestyle='-')
plt.rcParams['xtick.labelsize']=25
plt.rcParams['ytick.labelsize']=25
plt.legend(loc='upper left',prop={'size': 25})
plt.suptitle('Posts and Replies Graph hourly window', fontsize=30)

plt.yscale('log')

plt.show()
```

```python
density.diff().plot(figsize=(10,10), linewidth=5, fontsize=20)
```

```python

densityMonth = density['2016-08-01': '2016-08-15']
plt.rcParams['xtick.labelsize']=15
plt.rcParams['ytick.labelsize']=15
densityMonth.Density.plot(title='Zoom in August')
```

```python

densityDay = density['2016-08-13': '2016-08-15']
densityDay.Density.plot(title='Zoom in 13-15 August')
```

```python

densityHr = density['2016-08-14 01:00': '2016-08-15 01:00']
densityHr.Density.plot(title='Zoom in 14 August 01:00 to 15 August 01:00')
```
