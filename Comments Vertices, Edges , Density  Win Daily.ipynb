```python
import pandas as pd
from datetime import datetime
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

import numpy as np
```

# Posts and Replies Graph derived from sample file for daily Window Analysis


### Vertices and edges

```python
densityVandE=pd.read_csv('Density/densityWindow.csv')
densityVandE['Date'] = pd.to_datetime(densityVandE['Date'], format='%d/%m/%Y %H:%M:%S')
densityVandE.set_index('Date', inplace=True)
densityVandE=densityVandE[['TotalVertices','TotalEdges']]
densityVandE.plot(figsize=(30,10),style=['-', '--','.-'])
plt.grid(linestyle='-')
plt.rcParams['xtick.labelsize']=20
plt.rcParams['ytick.labelsize']=20
plt.legend(loc='upper left',prop={'size': 25})
plt.suptitle('Posts and Replies Graph daily window', fontsize=30)


plt.show()
```

### Zoom in to peak in Oct 10 th

```python

verticesMonth = densityVandE['2016-10-08': '2016-10-10']
verticesMonth.plot(style=['-', '--'])
verticesMonth.TotalVertices.plot(title='# of Vertices October')
plt.grid(linestyle='-')
plt.rcParams['xtick.labelsize']=15
plt.rcParams['ytick.labelsize']=15
plt.show()
```

# Density

```python
density=pd.read_csv('Density/densityWindow.csv')
density['Date'] = pd.to_datetime(density['Date'], format='%d/%m/%Y %H:%M:%S')
density.set_index('Date', inplace=True)
density=density[['Density']]
density.plot(figsize=(30,10),style=['-', '--','.-'])
plt.grid(linestyle='-')
plt.rcParams['xtick.labelsize']=20
plt.rcParams['ytick.labelsize']=20
plt.legend(loc='upper left',prop={'size': 25})
plt.suptitle('Posts and Replies Graph daily window', fontsize=30)
plt.yscale('log')


plt.show()
```
