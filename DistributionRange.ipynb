```python
import pandas as pd
from datetime import datetime
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

import numpy as np
```

```python
commentsDistrib=pd.read_csv('Distribution/distribRangeIn.csv')
commentsDistrib['Date'] = pd.to_datetime(commentsDistrib['Date'], format='%d/%m/%Y %H:%M:%S')
commentsDistrib.set_index('Date', inplace=True)
#distribution['2016-08-11']
monthComments=commentsDistrib['2016/10/09': '2016/10/09']
finalComments=monthComments[['InDegree','Total']]
xCom = finalComments.InDegree
yCom = finalComments.Total


commentsUser=pd.read_csv('Distribution/distribRangeInUsr.csv')
commentsUser['Date'] = pd.to_datetime(commentsUser['Date'], format='%d/%m/%Y %H:%M:%S')
commentsUser.set_index('Date', inplace=True)
#distribution['2016-08-11']
monthUsers=commentsUser['2016/10/09': '2016/10/09']
finalUsers=monthUsers[['InDegree','Total']]
xUsr = finalUsers.InDegree
yUsr = finalUsers.Total
```

```python
fig = plt.figure(figsize=(10,10))
plt.rcParams['xtick.labelsize']=15
plt.rcParams['ytick.labelsize']=15
ax1 = fig.add_subplot(111)
plt.plot(xCom,yCom,'.', markersize=20,label='Posts and Replies graph')
plt.grid(linestyle='-')
plt.plot(xUsr,yUsr,'s',linewidth=4,label='Posting and Replying Users graph')
fig.suptitle('Distribution', fontsize=20)
plt.xlabel('In-Degree', fontsize=18)
plt.ylabel('Count', fontsize=16)
plt.legend(loc='upper right',prop={'size': 15})
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
plt.xlabel('In-Degree', fontsize=18)
plt.ylabel('Count', fontsize=16)
fig.suptitle('Log-Log Scale Distribution', fontsize=20)

#ax1 = fig.add_subplot(111)
plt.plot(xCom,yCom,'.', markersize=20,label='Posts and Replies graph')
plt.grid(linestyle='-')
plt.plot(xUsr,yUsr,'s',linewidth=3,label='Posting and Replying Users graph')
plt.grid(linestyle='-')

plt.yscale('log')
plt.xscale('log')

plt.legend(loc='upper right',prop={'size': 15})

plt.show()
```
