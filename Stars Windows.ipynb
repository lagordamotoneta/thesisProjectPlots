```python
import pandas as pd
from datetime import datetime
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

import numpy as np
```

```python
stars=pd.read_csv('Stars/starsRange.csv')
cc=pd.read_csv('CC/CCRange.csv')
cc['Date'] = pd.to_datetime(cc['Date'], format='%d/%m/%Y %H:%M:%S')
cc.set_index('Date', inplace=True)
ccs=cc[['Size']]
ax=ccs.plot(figsize=(30,10),style=['-', '--','.-'])

stars['Date'] = pd.to_datetime(stars['Date'], format='%d/%m/%Y %H:%M:%S')
stars.set_index('Date', inplace=True)
starsDaily=stars[['InDegree']]
starsDaily.plot(figsize=(30,10),style=['-', '--','.-'],ax=ax)
# plt.grid(linestyle='-')
# plt.rcParams['xtick.labelsize']=25
# plt.rcParams['ytick.labelsize']=25
plt.legend(loc='upper left',prop={'size': 25})
# plt.suptitle('InDegree of identified star Post, accumulated', fontsize=40)

plt.show()
```

```python
stars=pd.read_csv('Stars/starsWindow.csv')
stars['Date'] = pd.to_datetime(stars['Date'], format='%d/%m/%Y %H:%M:%S')
stars.set_index('Date', inplace=True)
starsDaily=stars[['InDegree']]
starsDaily.plot(figsize=(30,10),style=['-', '--','.-'])
plt.grid(linestyle='-')
plt.rcParams['xtick.labelsize']=25
plt.rcParams['ytick.labelsize']=25
plt.legend(loc='upper left',prop={'size': 25})
plt.suptitle('InDegree of identified star Posts per day', fontsize=40)

plt.show()


```

```python
starsHr=pd.read_csv('Stars/starsWindowHr.csv')
starsHr['Date'] = pd.to_datetime(starsHr['Date'], format='%d/%m/%Y %H:%M:%S')
starsHr.set_index('Date', inplace=True)
starsHourly=starsHr[['InDegree']]
starsHourly.plot(figsize=(30,10),style=['-', '--','.-'])
plt.grid(linestyle='-')
plt.rcParams['xtick.labelsize']=25
plt.rcParams['ytick.labelsize']=25
plt.legend(loc='upper left',prop={'size': 25})
plt.suptitle('InDegree of identified star Posts per hour', fontsize=40)

plt.show()


```

```python
starsHourly.diff().plot(figsize=(30,10), linewidth=2, fontsize=20)
plt.legend(loc='upper left',prop={'size': 25})
plt.suptitle('Remove the trend InDegree of identified star Posts per hour', fontsize=40)
plt.grid(linestyle='-')
plt.show()
```
