```python
import pandas as pd
from datetime import datetime
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

import numpy as np
```

# Post and Replies Graph derived from sample file for Range Analysis

### Vertices and Edges

```python
densityVandE=pd.read_csv('Density/densityRange.csv')
densityVandE['Date'] = pd.to_datetime(densityVandE['Date'], format='%d/%m/%Y %H:%M:%S')
densityVandE.set_index('Date', inplace=True)
densityVandE=densityVandE[['TotalVertices','TotalEdges']]
densityVandE.plot(figsize=(20,10),style=['-', '--','.-'])
plt.grid(linestyle='-')
plt.rcParams['xtick.labelsize']=25
plt.rcParams['ytick.labelsize']=25
plt.legend(loc='upper left',prop={'size': 25})
plt.suptitle('Posts and Replies Graph', fontsize=30)
plt.yscale('log')

plt.show()
```

### Density

```python
densityVandE=pd.read_csv('Density/densityRange.csv')
densityVandE['Date'] = pd.to_datetime(densityVandE['Date'], format='%d/%m/%Y %H:%M:%S')
densityVandE.set_index('Date', inplace=True)
densityVandE=densityVandE[['Density']]
densityVandE.plot(figsize=(20,10),style=['-', '--','.-'])
plt.grid(linestyle='-')
plt.rcParams['xtick.labelsize']=20
plt.rcParams['ytick.labelsize']=20
plt.legend(loc='upper left',prop={'size': 25})
plt.suptitle('Posts and Replies Graph', fontsize=30)

plt.yscale('log')

plt.show()
```
